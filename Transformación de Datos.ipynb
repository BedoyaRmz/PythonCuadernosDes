### MANEJO NULL

```python
taxpayer = open('C:\\Users\\02ADSI\\Desktop\\estaciones eolicas\\files\\06\\Estacion1.csv', 'r').readlines()
taxpayer = [row[0:-1] for row in taxpayer]
taxpayer = [row.split(';') for row in taxpayer]
taxpayer
```

```python
taxpayer = [ [element if element != 'NULL' else None for element in row] for row in taxpayer]
taxpayer
```

```python
[row for row in taxpayer if row[1] is not None]
```

### Ordenamiento

```python
from operator import itemgetter
sorted([row for row in mail if int(row[2]) > 200], key=itemgetter(2))
```

```python
mail = [[element if index != 2 else int(element) for index, element in enumerate(row)] for row in mail]
mail
```

```python
sorted([row for row in mail if int(row[2]) > 350], key=itemgetter(3))
```

```python
sorted([row for row in mail], key=itemgetter(1,3))
```
